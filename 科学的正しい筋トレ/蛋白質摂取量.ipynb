# 紹介

2009年　カナダ・ロンドン大学
体重の異なる20代と70代を集め、レッグエクステンションを行った後に0~40gの蛋白質を摂取させ、筋蛋白質の合成率を計測。
合成率が最も高まる摂取量は「年齢」と「体重」によって違うことが分かった。

## 一食当りの最適な蛋白質摂取量

- 20代:体重1kg当たり平均0.24g (0.18g - 0.30g)
- 70代:体重1kg当たり平均0.40g (0.21g - 0.59g)

```python
import pandas as pd
```

```python
df = pd.DataFrame([*range(40,100,5)], columns=['体重'])
df['20代'] = df['体重'] * 0.24
df['70代'] = df['体重'] * 0.40
df
```

## 24時間の最適蛋白質摂取量

2017年　マクマスター大学　49の研究報告を元にしたメタアナリシス

体重1kg当たり平均1.62g (1.03g - 2.20g)

```python
df = pd.DataFrame([*range(40,100,5)], columns=['体重'])
df['摂取量/平均'] = df['体重'] * 1.62
df['摂取量/最小'] = df['体重'] * 1.03
df['摂取量/最大'] = df['体重'] * 2.20
df
```
